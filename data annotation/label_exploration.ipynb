```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
colors = {
    "contextual contradiction": "C0",
    "mock enthusiasm": "C1",
    "stylistic irony": "C2", 
    "institutional critique": "C3", 
    "behavioral observation": "C4"
}
```

```python
df = pd.read_csv('sarcasm_classifications.csv')
df.head()
```

```python
value_counts = df['classification'].value_counts()
labels = value_counts.index.to_numpy()
counts = value_counts.to_numpy()
```

```python
plt.pie(counts, labels=labels, colors=[colors[key] for key in labels], autopct='%1.1f%%')
plt.title("Distribution after initial labeling")
plt.show() 
```
